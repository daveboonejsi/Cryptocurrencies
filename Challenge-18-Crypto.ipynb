```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import pandas as pd
#import hvPlot.pandas
#import plotly.express as px
```

```python
file_path = "crypto_data.csv"
crypto_df = pd.read_csv(file_path, index_col=0)
crypto_df
```

```python
crypto_df["Algorithm"].isna().sum()
```

```python
#Remove non trading cryptocurrencies 
new_crypto_df = crypto_df[crypto_df['IsTrading']==True]
new_crypto_df
```

```python
# Remove cryptocurrencies with no algorithm defined
new_crypto_df[new_crypto_df['Algorithm']!="N/A"]
```

```python
#drop IsTrading column
new_crypto_df = new_crypto_df.drop(["IsTrading"], axis=1)
new_crypto_df
```

```python
# Remove cryptocurrencies with null value
new_crypto_df = new_crypto_df.dropna()
new_crypto_df
```

```python
# Remove cryptocurrencies with no coins mined
new_crypto_df = new_crypto_df[new_crypto_df['TotalCoinsMined']!=0]
new_crypto_df
```

```python
# Store cryptocurrency names in df
data = new_crypto_df["CoinName"]
coins_name = pd.DataFrame(data, index=new_crypto_df.index)
coins_name
```

```python
#Remove the CoinName column
newer_crypto_df= new_crypto_df.drop(['CoinName'], axis=1)
newer_crypto_df
```

```python
# Saving cleaned data
file_path = "/Users/dboone/Desktop/bootcamp/Cryptocurrencies/Crypto_data_cleaned.csv"
new_crypto_df.to_csv(file_path, index=False)
```

```python

X = pd.get_dummies(data=newer_crypto_df, columns=["Algorithm","ProofType"])
X
```

```python
#reset the index
X.reset_index(drop=True, inplace=True)
X
```

```python
# scale the data
X_scaled=StandardScaler().fit_transform(X)
X_scaled
```

```python
#use PCA model to reduce dimensions of X
pca=PCA(n_components=3)
X_pca= pca.fit_transform(X_scaled)
```

```python
#Transform PCA data into dataframe and Add the index column again.
pcs_df = pd.DataFrame(
    data=X_pca, columns=["PC1","PC2","PC3"], index=coin_name.index
)
pcs_df
```

```python
#get explained variance
pca.explained_variance_ratio_
```

```python
# Find optimum value of K
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initialize and fit K-means model 
model = KMeans(n_clusters=4, random_state=0)
model.fit(pcs_df)
# Predict clusters
predictions = model.predict(pcs_df)
# Add the predicted class columns
pcs_df["Class"] = model.labels_
pcs_df.head()
```

```python
model.cluster_centers_
```
